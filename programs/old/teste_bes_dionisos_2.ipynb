```python
# Importar dependências
import gempy as gp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

import warnings
warnings.filterwarnings("ignore")

pd.set_option('precision', 2)

# Criando o objeto modelo e variável para indicar a pasta com os dados
data_path = '../input/Teste_Dionisos/'  
geo_model = gp.create_model('BES_DIONISOS_GRID')
```

```python
gp.init_data(geo_model,
             extent=[405000, 575000, 7655000, 7800000, -9777, -795],
             resolution=[200, 200, 100],
             path_i=data_path + "sp_lessFM_2.csv")
```

```python
# Ordenando as surfaces (Estratigrafia (topo para base)) 1
gp.map_stack_to_surfaces(geo_model,                         
                         {"FM0_series": ('FM0'),
                          "FM6_series": ('FM6'),
                          "FM12_series": ('FM12'),
                          "FM18_series": ('FM18'),
                          "FM24_series": ('FM24'),
                          "base_series": ('basement')})

# Ver o grid do modelo
geo_model.grid
```

```python
geo_model.surfaces
```

```python
# Adicionando orientações manualmente (peguei a mediana dos valores de FM0)
geo_model.add_orientations(X=465000, Y=7800000, Z=-6225, surface='FM0', pole_vector=(0,0,1))
```

```python
# Ver em 2D os dados nas 3 direções
gp.plot_2d(geo_model, direction=['x'], show_data=True)

gp.plot_2d(geo_model, direction=['y'], show_data=True)

gp.plot_2d(geo_model, direction=['z'], show_data=True)
```

```python
# Ver os dados em 3D
gpv = gp.plot_3d(geo_model, image=False, plotter_type='basic', show_data=True)
```

```python
%%time
# Interpolador
gp.set_interpolator(geo_model,
                    theano_optimizer='fast_run',
                    compile_theano=True
                    )
```

```python
%%time
# Computar a solução do modelo
sol = gp.compute_model(geo_model, compute_mesh=True)
```
