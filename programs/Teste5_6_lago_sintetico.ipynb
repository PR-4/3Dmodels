```python
import gempy as gp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
data_path = '../input/Teste5/'
geo_model = gp.create_model('SyntheitcBasinTeste5')
```

```python
gp.init_data(geo_model,
             extent=[0, 2000, -2000, 2000, -100, 1200],
             resolution=[50, 50, 50],
             path_i=data_path + "6_lago_sintetico/surfaces_points.csv")
```

```python
geo_model.surfaces
```

```python
gp.map_stack_to_surfaces(geo_model,
                         {"Strat_Series": ('Formacao_1', 'Formacao_2', 'Formacao_3', 'Formacao_4')})
```

```python
geo_model.surface_points
```

```python
geo_model.orientations
```

```python
geo_model.add_orientations(X=600, Y=500, Z=550, surface="Formacao_2", pole_vector=(0,0,1))
```

```python
geo_model.grid
```

```python
plot = gp.plot_2d(geo_model, show_lith=False, show_boundaries=False)
plt.show()
```

```python
gpv = gp.plot_3d(geo_model, image=False, plotter_type='basic')
```

```python
gp.set_interpolator(geo_model,
                    compile_theano=True,
                    theano_optimizer='fast_compile',
                    )
```

```python
sol = gp.compute_model(geo_model)
sol
```
