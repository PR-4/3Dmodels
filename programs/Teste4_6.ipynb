```python
import gempy as gp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
data_path = '../input/Teste4/'
geo_model = gp.create_model('SyntheitcBasinTeste4')
```

```python
gp.init_data(geo_model,
             extent=[0, 550, 0, 550, -500, 550],
             resolution=[50, 50, 50],
             path_i=data_path + "surface_points.csv",
             path_o=data_path + "orientations_points.csv")
```

```python
geo_model.surfaces
```

```python
gp.map_stack_to_surfaces(geo_model,
                          {"Sedimentary_Series": ('A', 'B', 'C', 'basement'), 
                           })
```

```python
geo_model.grid
```

```python
plot = gp.plot_2d(geo_model, show_lith=False, show_boundaries=False)
plt.show()
```

```python
gp.set_interpolator(geo_model,
                    compile_theano=True,
                    theano_optimizer='fast_compile',
                    )
```
