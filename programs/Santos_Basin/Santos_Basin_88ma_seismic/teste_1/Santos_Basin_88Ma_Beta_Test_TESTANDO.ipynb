```python
# Importing dependency
import gempy as gp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.precision', 2)

# Data path, creating the model object and the name
data_path = '../../../../input/Santos_Basin/Santos_Basin_88ma_seismic/v1.0.0/'
nome_modelo = "Santos_Basin_88ma_Teste_1"
geo_model = gp.create_model(nome_modelo)
```

```python
# Extension of the model, resolution of the model and the paths of surface and orientation
gp.init_data(
    geo_model,
    extent=[0, 4000, 0, 42500, -13000, 0],
    resolution=[50, 50, 50],
    path_i=data_path + "surfaces_points.csv",
    #path_o=data_path + "orientations_points.csv"
    )
```

```python
# The surfaces of the model
gp.get_data(geo_model, 'surfaces')
```

```python
# Sorting the surfaces and defining a serie for them (Stratigraphy - Top to Bottom)
gp.map_stack_to_surfaces(
    geo_model,
    {
        "Strat_Series_1": ('CAMBURI_II'),
        "Strat_Series_2": ('CAMBURI_I'),
        "Strat_Series_3": ('ARIRI'),
        "Fault_series": ('FALHA'),
        "Strat_Series_4": ('BV_II'),
        "Strat_Series_5": ('BV_I'),
        "Strat_Series_6": ('ITAPEMA'),
        "Strat_Series_7": ('CAMBORIU'),
        "Basement": ('basement'),
        }
    )

# Ver o grid do modelo
geo_model.grid
```

```python
# Declaring the fault serie
geo_model.set_is_fault(['Fault_series'])
geo_model.set_is_finite_fault(series_fault=['Fault_series'])
```

```python
# The surfaces of the model
gp.get_data(geo_model, 'surfaces')
```

```python
geo_model.series
```

```python
neighbours = gp.select_nearest_surfaces_points(geo_model, geo_model.surface_points.df, 5)
opoints = gp.set_orientation_from_neighbours_all(geo_model, neighbours)

# Salvando os pontos de orientação gerados pelo KNN (K=10)
orientations_p = geo_model.orientations.df
orientations_p.to_csv(data_path + "orientations_points.csv")

# Trocando o nome de surface para formation (necessário se for importar no Gempy)
orientations_p_formation = orientations_p.copy()
orientations_p_formation = orientations_p_formation.rename(columns={'surface': 'formation'})
orientations_p_formation.to_csv(data_path + "orientations_points_2.csv")
orientations_p_formation
```
