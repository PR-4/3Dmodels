## Negativando os valores de z

```python
def valorneg(valor):
    return -valor

surface_cln['Z'] = surface_cln['Z'].apply(valorneg)
print(surface_cln)
```

## Salvando novo csv

```python
surface_cln.to_csv('sp_clean.csv', index=False)
```

## Valores máximos e mínimos

```python
surface_cln = pd.read_csv('sp_clean_2.csv', delimiter=',')
```

```python
max_x = surface_cln['X'].max()
min_x = surface_cln['X'].min()
max_y = surface_cln['Y'].max()
min_y = surface_cln['Y'].min()
max_z = surface_cln['Z'].max()
min_z = surface_cln['Z'].min()

print("Valor máximo de X: ", max_x)
print("Valor mínimo de X: ", min_x)
print("Valor máximo de Y: ", max_y)
print("Valor mínimo de Y: ", min_y)
print("Valor máximo de Z: ", max_z)
print("Valor mínimo de Z: ", min_z)
```

```python
mean_values = surface_cln.groupby('formation')['X', 'Y', 'Z'].mean()
```

```python
mediana_por_formacao = surface_cln.groupby('formation').median()
```

```python
mediana_por_formacao
```

```python
remove_fm = ["FM1", "FM2", "FM3", "FM4", "FM5", "FM7", "FM8", "FM9", "FM10", "FM11", "FM13", "FM14", "FM15", "FM16", "FM17", "FM19", "FM20", "FM21", "FM22", "FM23"]
```

```python
mask = surface_cln["formation"].isin(remove_fm)
```

```python
df_fm = surface_cln[~mask]
```

```python
df_fm
```

```python
df_fm.to_csv('sp_lessFM.csv', index=False)
```
